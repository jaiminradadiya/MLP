```python
data = {'x':range(1,12),
        'y' : [5,1,2,3,5,4,6,7,8,9,100]}

df = pd.DataFrame(data)
```

```python
plt.plot(df['x'], df['y'], marker='^', color='red')
plt.show()
```

```python
plt.plot(df['x'], df['y'], marker='.', color='red')
plt.xlabel( 'X-axis-Days')
plt.ylabel('Y-axis-salesValue ')
plt.title('SALES REPORT - DayWise')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['x'], df['y']),
plt.xlabel( 'bar')
plt.ylabel('Y-axis-salesValue ')
plt.title('SALES REPORT - DayWise')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['x'], df['y']),
plt.xlabel( 'bar')
plt.ylabel('Y-axis-salesValue ')
plt.title('SALES REPORT - DayWise')
plt.grid(True)
plt.show()
```

```python
#Pie Chart
# Sample DataFrame
data = { 'labels': ['A', 'B', 'C', 'D', 'E'], 'sizes': [15, 30, 25, 10, 20]}
df = pd.DataFrame(data)

# Pie chart
plt.pie(df['sizes'], labels=df['labels'], autopct='%1.1f%%'), plt.title('Pie Chart')
plt.title('Princes piechart')
plt.show()
```
