```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
np.sin(np.pi/2.)
```

```python
x = np.linspace(-np.pi, np.pi, 201)
plt.plot(x, np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.show()
```

```python
x = np.arange(0.75,5,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.array([0,6])
y = np.array([0,250])
plt.plot(x,y)
plt.show()
```

```python
x = np.array([0,6])
y = np.array([0,250])
plt.plot(x,y,'o')
plt.show()
```

```python
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```

```python
# marker in graph o ,*,+
# https://www.w3schools.com/python/matplotlib_markers.asp
y = np.array([3,8,1,10,5,7])
plt.plot(y,marker = '|')
plt.show()
```

```python
# Format Strings fmt
# marker|line|color
# https://www.w3schools.com/python/matplotlib_markers.asp
y = np.array([3,8,1,10])
plt.plot(y,'o:y')
plt.show()
```

```python
# Set the size and color of the marker
y = np.array([3,8,1,10])
plt.plot(y,marker = 'o', ms= 15, mec='b', mfc='hotpink')
plt.show()
```

**Linestyle**


1.   dotted

3.   dashdot






```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashdot')
plt.show()
```

**Line Color**

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```
