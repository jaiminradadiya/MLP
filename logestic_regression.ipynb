```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv('TeamsHalf.csv')
data.head()
```

```python
# Dropping Null Values
# data.dropna(inplace=True)
X = data.drop(columns=['rank'])  # Features
y = data['rank']
```

```python
categorical_features = ['lgID', 'tmID']
```

```python
from sklearn.preprocessing import LabelEncoder
preprocessor = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(), categorical_features)],
    remainder='passthrough'
)
X_encoded = preprocessor.fit_transform(X)
```

```python
sliced_y = data.iloc[:,3]
sliced_y
```

```python
train_x, test_x, train_y, test_y = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
reg = LogisticRegression()
reg.fit(train_x, train_y)
```

```python
# est_x_encoded = test_x_encoded[train_x_encoded.columns]
predictions = reg.predict(test_x)
```
